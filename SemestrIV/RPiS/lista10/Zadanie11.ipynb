```python
import matplotlib.pyplot as pyplot
import numpy
```

Na poczÄ…tek zapoznanie z danymi:

```python
x = [1,3,4,6,8,9,11,14]
y = [1,2,4,4,5,7,8 ,9]

pyplot.scatter(x,y);
```

Chcemy znaleÅºÄ‡ prostÄ… regresji,czyli prostÄ… w postaci
$$
y = \beta_0 + \beta_1 \cdot x
$$

Wektor wspÃ³Å‚czynnikÃ³w $\beta$ moÅ¼emy policzyÄ‡ korzystajÄ…c ze wzoru z wykÅ‚adu:
$$
\beta = (X^t X)^{-1}X^tY,
$$
gdzie $Y$ to wektor y'kÃ³w, a $X$ to macierz w postaci 

$$
X = \left[
    \begin{array}{cc}
        1 & x_1 \\
        1 & x_2 \\
        \vdots & \vdots \\
        1 & x_n
    \end{array}
\right]
$$

```python
X = []
for xi in x:
    X.append([1,xi])
Y = y

X = numpy.array(X)
Y = numpy.array(Y)
```

```python
beta = X.T.dot(X)
beta = numpy.linalg.inv(beta)
beta = beta.dot(X.T)
beta = beta.dot(Y)

print(beta)
```

```python
b_0 = beta[0]
b_1 = beta[1]

def f(x):
    return b_0 + b_1 * x

def map_list(f, x): 
    return list(map(f,x))

linspace = numpy.linspace(1,14);

pyplot.scatter(x,y,color="Red")
pyplot.plot(linspace, map_list(f, linspace))
pyplot.show()
```
